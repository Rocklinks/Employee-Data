#Draw a Line Plot to show the Number of the years worked in company and the Perfroamce Rating for that

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
sns.lineplot(x='YearsAtCompany', y='PerformanceRating', data=df,color='r')
plt.title("Employee's Years of Experience and their performance Rating")
plt.show()
```

##Observation: From the above line plotwe can see that the Employees those who worked nearly or equal to 30 years has the High performance Rating

#Find Joblevel vs Monthly Income using Bar graph

```python
plt.title('Monthly Income for Differnt Job Levels')
sns.barplot(x='JobLevel', y='MonthlyIncome', data=df,color='yellow')
plt.show()
```

##Observation: From the above bar graph we can see that difference of monthly incomes for different Job Levels





#Find Monthly Income Department wise

```python
plt.figure(figsize=(6,6))
plt.title('Monthly Income Department Wise')
sns.barplot(x='Department', y='MonthlyIncome', data=df,color='green')
plt.show()
```

##Observations: From the above observation we can see the Monthly Salary of the Employees Department Wise

#Find the Average Age of the Employees and the count of Employees on that Age the Histogram

```python
plt.figure(figsize=(6,8))
plt.title('Employee Age Distribution')
sns.histplot(df['Age'],color='cyan')
plt.vlines(df['Age'].mean(),ymin=0,ymax=470,colors='red',ls='--')
plt.show()
```

##Observations: From the Histogram we can find the number of employees at particular age. We also know the average age of the employee

#Find the Marital Status of the employees

```python
plt.figure(figsize=(6,6))
plt.title('Marital Status of Employees')
plt.pie(df['MaritalStatus'].value_counts(), autopct='%1.1f%%',labels=df['MaritalStatus'].value_counts().index,startangle=140)
plt.show()
```

##From the Above pie chart we can see the Percentage of the Employees who are married,single and divorced

#JobSatisfaction vs. PerformanceRating of Gender

```python
plt.title('Job Satisfaction Vs Performance Rating of Gender')
sns.scatterplot(x='JobSatisfaction', y='PerformanceRating', hue='Gender',data=df)
plt.show()
```

##Observations: From the Above Scatterplot we can find the Job Satisfaction vs Performance Rating of the gender clearly

#Create a Heatmap using monthly income,job satisfaction,Totoal working years and age by Department and Gender

```python
plt.figure(figsize=(12,8))
plt.title('Distribution of Monthly Income, Job Satisfaction, Total Working Years, and Age by Department and Gender')
pivot=df.pivot_table(values=['MonthlyIncome','JobSatisfaction','TotalWorkingYears','Age'],index='Department',columns='Gender')
sns.heatmap(pivot,cmap='Blues',annot=True)
plt.xlabel('Monthly Income, Job Satisfaction, Total Working Years and Age by Gender')
plt.yticks(rotation=0)
plt.show()
```

##Observation from the Above Observation we can find the Employees income,satisfaction,working hours and age by gender and Department

# Find Employee monthly income grouped by department and job role using plot.express

```python

fig = px.treemap(df, path=["Department", "JobRole"], values="MonthlyIncome",
                  color="PerformanceRating",
                  title="Employee Monthly Income by Department and Job Role")
fig.show()
```

##Observations: From the Above Tree Map we can clearly see the Monthly Income of an Employee grouped By Department and job Role
